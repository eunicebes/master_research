```python
import os
from collections import defaultdict
```

```python
recent_tweets_file = [filename for filename in os.listdir("Bipolar Tweets since March")]
recent_user_dict = {}
for file in recent_tweets_file:
    with open("Bipolar Tweets since March/"+file, "r") as data:
        records = data.readlines()
        for record in records:
            recent_user_dict[record.split("\t")[1]] = record.split("\t")[0]
len(recent_user_dict)
```

```python
user_dict = {}
with open("Bipolar Tweets by keyword Diagnosed Bipolar/bipolar_user_timezone", "r") as data:
    records = data.readlines()
    for record in records:
        user_dict[record.split("\t")[1]] = record.split("\t")[0]
len(user_dict)
```

```python
out_folder = "Bipolar Tweets with user ID"
updated_user_list = []
tweets_file = [filename for filename in os.listdir("Bipolar Tweets by keyword Diagnosed Bipolar/patient_tweets")]
for file in tweets_file:
    output_content = ''
    try:
        userid = user_dict[file]
        with open("Bipolar Tweets by keyword Diagnosed Bipolar/patient_tweets/"+file) as f:
            data = f.readlines()
            renewed_data = [userid+"\t"+file+"\t"+"\t".join(d.split("\t")[2:]) for d in data]
            checked_data = [d.split("\t")[2:] for d in data]
        f.close()
        try:
            with open("Bipolar Tweets since March/"+userid) as f:
                recent_data = f.readlines()
                # Check if the tweet is duplicate or not
                filtered_data = [d for d in recent_data if d.split("\t")[2:] not in checked_data]
                joined_data = filtered_data + renewed_data
            f.close()
            updated_user_list.append(userid)
            with open(out_folder + '/' + userid,'a') as output_file:
                output_content = ''.join(joined_data)
                output_file.write(output_content)
            output_file.close()
        except:
            continue
            with open(out_folder + '/' + userid,'a') as output_file:
                output_content = ''.join(renewed_data)
                output_file.write(output_content)
            output_file.close()
    except:
        continue
```

```python
output_content = ""
with open("Bipolar Tweets by keyword Diagnosed Bipolar/bipolar_user_timezone", "r") as data:
    records = data.readlines()
    for record in records:
        if record.split("\t")[0] in updated_user_list:
            output_content += record
with open("Bipolar Tweets by keyword Diagnosed Bipolar/lost_user", "a") as output_file:
    output_file.write(output_content)
```
