## Get user tweets

```python
def read_file(folder, filename):
    with open(folder + filename, 'r') as openfile:
        return [line.strip().split('\t') for line in openfile.readlines()]
```

```python
folder = "../tweet_process/Bipolar Tweets with user ID/"
user_file = [filename for filename in os.listdir(folder)]
patient_tweets_dict = defaultdict(lambda : [])
for userid in user_file:
    patient_tweets_dict[userid] = read_file(folder, userid)
print("Total patients in tweets folder: " + str(len(patient_tweets_dict)))
```

## Predict gender

```python
def predict_gender(users):
    user_gender = defaultdict()
    for user in users:
        texts = ""
        for text in user[1]:
            texts += text + "\n"
        gender = predictor.get_gender(texts)
        user_gender[user[0]] = gender
    return user_gender
```

```python
# function to delete url
def del_url(line):
    return re.sub(r'(\S*(\.com).*)|(https?:\/\/.*)', "", line)
# replace tag
def checktag(line): 
    return re.sub(r'\@\S*', "", line)
```

```python
users_tweets_content = []
for userid, data in patient_tweets_dict.items():
    each_content_list = []
    for tweet in data:
        try:
            each_content_list.append(del_url(checktag(tweet[3])))
        except:
            continue
    users_tweets_content.append((userid, each_content_list))
```
